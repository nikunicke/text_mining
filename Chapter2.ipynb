```python
target = ['men', 'women', 'people']
cfd = nltk.ConditionalFreqDist(
        (fileid, word)
        for fileid in state_union.fileids()
        for w in state_union.words(fileid)
        for word in target
        if w.lower().startswith(word) and re.match('men[\w]+', w.lower()) is None)
years = list(state_union.fileids())
cfd.tabulate(conditions = years, samples = target)
```

```python
def meronyms_holonyms(noun):

    part_meronyms = []
    substance_meronyms = []
    member_meronyms = []
    
    part_holonyms = []
    substance_holonyms = []
    member_holonyms = []
    
    
    for ss in wn.synsets(noun):
        
#         meronyms
        part_meronyms.append(ss.part_meronyms())
        substance_meronyms.append(ss.substance_meronyms())
        member_meronyms.append(ss.member_meronyms())
        
#         holonyms
        part_holonyms.append(ss.part_holonyms())
        substance_holonyms.append(ss.substance_holonyms())
        member_holonyms.append(ss.member_holonyms())
        
    print('Part meronyms: ', part_meronyms, '\n\nSubstance meronyms: ', substance_meronyms, '\n\nMember meronyms: ', member_meronyms, '\n\nPart holonyms: ', part_holonyms, '\n\nSubstance holonyms: ', substance_holonyms, '\n\nMember holonyms: ', member_holonyms)
        
```

```python
def supergloss(s):
    print('SYNSET DEFINITION:\n', wn.synset(s).definition())
    print('\nHYPERNYMS: ')
    for hyper in wn.synset(s).hypernyms():
        print(hyper.definition())
        
    print('\nHYPONYMS: ')
    for hypo in wn.synset(s).hyponyms():
        print(hypo.definition())
            
    
    
```

I do not think it makes any sense that corpus.stopwords do not include punctuation or other special symbols. Without filtering/cleaning, the bigrams would mostly be pairs consisting of either on or two special characters.

```python
 def top_bigrams(text, number):
    stopword = stopwords.words('english')
    content = [w.lower() for w in text if w.lower() not in stopword]       

    print(len(list(nltk.bigrams(content))))
    print(len(set(list(nltk.bigrams(content)))))

    bigrams = list(nltk.bigrams(content))

    noSymb = []
    for asd in bigrams:
        list_to_string = ' '.join(asd)
        if re.search('[^\w\s]', list_to_string) is None:
            noSymb.append(tuple(list_to_string.split()))

    counter = {}
    for pair in noSymb:
        if pair in counter:
            counter[pair] += 1
        else:
            counter[pair] = 1


    freq_bigrams = sorted(counter, key=counter.get, reverse = True)

    top = freq_bigrams[:number]

    return top
    
```

Formatting word pairs from nltk.org/book/ch02.html exercise 28.

If you want to check with any other words. Input as following:

    'pair1-pair1, pair2-pair2, pair3-pair3...'

```python
def sample_data(string):
    refined_sample = re.sub('-', ',', re.sub('[\s]', '', string))
    sample_list = refined_sample.split(',')

    
    sample = []

    for i in range(0, 59, 2):
        sample.append(tuple([sample_list[i], sample_list[i + 1]]))

    return sample
    
```

```python
sample = sample_data('car-automobile, gem-jewel, journey-voyage, boy-lad, coast-shore, asylum-madhouse, magician-wizard, midday-noon, furnace-stove, food-fruit, bird-cock, bird-crane, tool-implement, brother-monk, lad-brother, crane-implement, journey-car, monk-oracle, cemetery-woodland, food-rooster, coast-hill, forest-graveyard, shore-woodland, monk-slave, coast-forest, lad-wizard, chord-smile, glass-magician, rooster-voyage, noon-string')
sample
```

```python
results = {}

for item in sample:
    ss1 = wn.synsets(item[0])
    ss2 = wn.synsets(item[1])

#     ss1 = wn.synsets(item[0])
#     ss2 = wn.synsets(item[1])
    
    score = round(ss1[0].path_similarity(ss2[0]), 3)
    
    results[item[0] + '-' + item[1]] = score
```

```python
plt.figure(figsize=(50,10))
plot_list = sorted( ((k, v) for v, k in results.items()), reverse = True)
x, y = zip(*plot_list)

plt.bar(y, x)
plt.show()


for i in plot_list:
    print(i[1], ': ', i[0])
```
