### EXERCISE 22
We start with creating an empty list that we will fill with words with the length equals to four. We iterate trough the input text and check add all words with the length of four to our new list. fDist will then count and sort our words in a decending order.

```python
def four_letters(text):
    
    len4_words = []
    
    for word in text:
        if len(word) == 4:
            len4_words.append(word)
    
    fdist = FreqDist(len4_words)
       
    print(fdist)
    return fdist.most_common()
            
```

### EXERCISE 28
We assign a variable that will count the how many times the given word occurs in the text. Then a for-loop will iterate trough every word in the text and if a word matches the given word, our counter will increase by one.

```python
def percent(word, text):
    
    word_occurence = 0
    
    for w in text:
        w.lower()
        if w == word.lower():
            word_occurence += 1
    
    percent = round((word_occurence / len(text) * 100), 2)
    
    print('Your word was found', word_occurence, 'times. That is equal to ', percent, '% of the total amount of words.')
```

```python
percent('whale', text1)
```
